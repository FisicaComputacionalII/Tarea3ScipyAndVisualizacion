## Teoría

### Sistema Masa-Resorte amortiguado-forzado

El sistema masa-resorte es gobernado por la ley de Hooke. En el caso de un resorte que oscila en una sola dimensión, si se tiene en cuenta la fricción (con constante $\gamma$) y un forzamiento armónico del tipo $F(t)=F_0\cos\left(\omega t\right)$, la ecuación diferencial del movimiento es 
	
$$\ddot{x}+2\beta\dot{x}+\omega_{0}^{2}x=\bar{F_{0}}\cos\left(\omega t\right)$$
	
donde $\beta=\frac{\gamma}{2m}$ (constante de amortiguamiento), $\omega_0=\sqrt{\frac{k}{m}}$ (frecuencia propia del oscilador) y $\bar{F_{0}}=\frac{F_{0}}{m}$.

Recuerde que para desacoplar la ecuación del oscilador, puede definir $\overrightarrow{x}=\begin{pmatrix}x\\\dot{x}\end{pmatrix}$, con lo que se transforma en dos ecuaciones de primer orden (muestre que es así)

$$ \dot{\overrightarrow{x}}=\begin{bmatrix} 0 & 1 \\ -\omega_{0}^{2} & -2\beta \end{bmatrix} \overrightarrow{x} + \begin{pmatrix}0\\ F_0\cos(\omega t)\end{pmatrix}$$
